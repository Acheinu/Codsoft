```python
# import libraries

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE 
```

```python
# modeling
X = Credit.drop('Class', axis = 1)
y = Credit['Class']
```

```python
# feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

 Handling class imbalance
 SMOTE ( oversampling the minority class)

```python
smote = SMOTE(random_state = 42)
X_train_res, y_train_res = smote.fit_resample(X_scaled, y)
```

Data Splitting

```python
# Split data into training and testing sets (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(X_train_res, y_train_res, test_size=0.2, random_state=42)
```

Training the model

1) Logistic Regression

```python
# initialize model
model_lg = LogisticRegression()

model_lg.fit(X_train, y_train)
```

```python
# model evaluation

y_pred = model_lg.predict(X_test)
y_predTrain=model_lg.predict(X_train)
```

```python
# evaluating models performance

print("\nAccuracy Score:")

print(f"Train Accuracy: {accuracy_score(y_train, y_train)}")

print(f"Test Accuracy: {accuracy_score(y_train, y_predTrain)}")
```

```python
# evaluating models performance

print("\nRecall Score:")

print(f"Train Recall: {recall_score(y_train, y_train)}")

print(f"Test Recall: {recall_score(y_train, y_predTrain)}")
```

```python
# making prdeictions
y_pred = model_lg.predict(X_test)

# evaluating models performance
accuracy1 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Make predictions
y_pred = model_lg.predict(X_test)  # or rf.predict(X_test) for Random Forest

# Generate classification report (Precision, Recall, F1-Score)
print(classification_report(y_test, y_pred))

# Confusion matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# confusion matrix

sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fraud', 'Fraud'], 
            yticklabels=['Not Fraud', 'Fraud'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
